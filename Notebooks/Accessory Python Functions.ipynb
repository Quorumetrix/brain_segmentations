```python
import os
import shutil
```

```python
'''
#Example filenames
341880_413020_002040 # Original file identifier
341880_413020_002040-DTI
341880_413020_002040-MRI
341880_413020_002040-Segmentation
'''


ROOT = 'Z://Collaboration_data/Iordonova_lab/my_generated/Iordanova_Control-M-R8_NeuN-cFOS/'
quicknii_slice_output = ROOT + 'temp/slices/'
neur_folder_path = 'Z://Collaboration_data/Iordonova_lab/my_generated/Iordanova_Control-M-R8_NeuN-cFOS/neun_ds/'
cfos_folder_path = 'Z://Collaboration_data/Iordonova_lab/my_generated/Iordanova_Control-M-R8_NeuN-cFOS/cfos_ds/'

filetype = '.png'



output_folder = 'M://Brain_Registration/continuous_output/'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Get a list of all files in the folder
input_folder = neur_folder_path
all_files = os.listdir(input_folder)

# Filter and sort only the files with the specified filetype
sorted_files = sorted([f for f in all_files if f.endswith(filetype)])

print(sorted_files)
```

```python
# Iterate through the sorted files and generate new filenames with continuous numbering
for index, filename in enumerate(sorted_files, start=1):

    # Copy and rename the original downsampled file (png)
    print(f"Current file: {filename}")
    new_filename = f"neun_ds_{index:04d}{filetype}"
    print(f"New file: {new_filename}")
    
    src_path = os.path.join(neur_folder_path, filename)
    dst_path = os.path.join(output_folder, new_filename)
    
    shutil.copy(src_path, dst_path)
    
    src_path = os.path.join(cfos_folder_path, filename)
    new_filename = f"cfos_ds_{index:04d}{filetype}"
    dst_path = os.path.join(output_folder, new_filename)
    
    shutil.copy(src_path, dst_path)  
    

    for ext in ['MRI', 'Segmentation', 'DTI']:

        src_path = os.path.join(quicknii_slice_output, filename[:-4]+'-'+ext+'.png')
        new_filename = f"{ext}_{index:04d}{filetype}"
        dst_path = os.path.join(output_folder, new_filename)
        shutil.copy(src_path, dst_path)        
```

```python
import os
import shutil

# source_folder = 'path/to/source/folder'
# destination_folder = 'path/to/destination/folder'

source_folder = 'H://Renders/Blender/DBC_Hoffman/neuromorpho_firing/'
destination_folder =  'M://Brain_Registration/test_copies1/'

target_file_count = 500

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# List all files in the source folder
file_list = os.listdir(source_folder)

# Filter only '.tif' files
tif_files = [f for f in file_list if f.endswith('.png')]

total_files = len(tif_files)

# Calculate the value of N to get the target number of files
N = max(total_files // target_file_count, 1)

# Sort the list of files
sorted_tif_files = sorted(tif_files)

# Copy every Nth file to the destination folder
for i, file in enumerate(sorted_tif_files):
    if i % N == 0:
        source_path = os.path.join(source_folder, file)
        destination_path = os.path.join(destination_folder, file)
        shutil.copy2(source_path, destination_path)

print(f"Files copied successfully. N={N}")
```
